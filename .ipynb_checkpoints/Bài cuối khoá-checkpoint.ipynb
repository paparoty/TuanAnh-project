```python
from tkinter import *
windows = Tk()
windows.title("QUẢN LÝ VẬT TƯ CÔNG TRƯỜNG. Author: TRẦN TUẤN ANH")
windows.geometry("1100x600")
Manager_dict= {}
material_dict= {}

   
# TẠO HÀM

def clbtsearch():
    print (Manager_dict)
    
def clbtcreate():
    for key in material_dict.keys():
        display_listbox.insert(key)


def clbtimport():
    material_dict['type'] = imptype_entry.get()
    material_dict['impQuantity'] = float(impQua_entry.get())
    material_dict['imptime'] = imptime_entry.get()
    material_dict['expiry'] = exp_entry.get()
    Manager_dict['material']=material_dict
    display_listbox.insert(Manager_dict)
def clbtexport():
    material_dict['type'] = exptype_entry.get()
    material_dict['expQuantity'] = float(expQua_entry.get())
    material_dict['exptime'] = exptime_entry.get()
    inventory_label1 = material_dict['impQuantity'] - material_dict['expQuantity']
    material_dict['inventory'] = inventory_label1
    Manager_dict['material']=material_dict
    display_listbox.insert(Manager_dict)

# NHẬP VÀ THAO TÁC
# LABEL
title_label = Label(windows, text = "QUẢN LÝ VẬT TƯ CÔNG TRÌNH",font=('arial',20,'bold')).grid(pady=(10,20),padx=(200,50),row=0,column=0,columnspan=5)
type_label = Label(windows, text = "CHỦNG LOẠI:",font=('arial',12,'bold')).grid(pady=(5,20),padx=(0,5),row=1,column=0)
imp_label = Label(windows, text = "VẬT TƯ NHẬP:",font=('arial',12,'bold')).grid(row=2,column=0)
impQuan_label = Label(windows, text = "SỐ LƯỢNG NHẬP:",font=('arial',12,'bold')).grid(row=2,column=2)
imptime_label = Label(windows, text = "THỜI GIAN NHẬP:",font=('arial',12,'bold')).grid(row=3,column=0)
exp_label = Label(windows, text = "HẠN SỬ DỤNG:",font=('arial',12,'bold')).grid(pady=(5,20),padx=(0,5),row=3,column=2)

exptype_label = Label(windows, text = "VẬT TƯ XUẤT:",font=('arial',12,'bold')).grid(row=5,column=0)
expQuan_label = Label(windows, text = "SỐ LƯỢNG XUẤT:",font=('arial',12,'bold')).grid(row=5,column=2)
site_label = Label(windows, text = "CÔNG TRÌNH:",font=('arial',12,'bold')).grid(row=6,column=2)
exptime_label = Label(windows, text = "THỜI GIAN XUẤT:",font=('arial',12,'bold')).grid(row=6,column=0)

# BUTTON
search_button = Button(windows,bd=8,font=('arial',10,'bold'),text='TÌM KIẾM',bg='silver', command = clbtsearch)
search_button.grid(row=1,column=2,sticky=N)
create_button = Button(windows,bd=8,font=('arial',10,'bold'),text='TẠO VẬT LIỆU MỚI',bg='silver', command = clbtcreate)
create_button.grid(row=1,column=4,sticky=N)
import_button = Button(windows,bd=8,font=('arial',10,'bold'),text='NHẬP',bg='silver', command = clbtimport)
import_button.grid(row=2,column=5)
export_button = Button(windows,bd=8,font=('arial',10,'bold'),text='XUẤT',bg='silver', command = clbtexport)
export_button.grid(row=5,column=5)

# ENTRY
type_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
type_entry.grid(row=1,column=1,sticky=N)
imptype_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
imptype_entry.grid(row=2,column=1,sticky=N)
impQua_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
impQua_entry.grid(row=2,column=4,sticky=N)
imptime_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
imptime_entry.grid(row=3,column=1,sticky=N)
exp_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
exp_entry.grid(row=3,column=4,sticky=N)

exptype_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
exptype_entry.grid(row=5,column=1)
expQua_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
expQua_entry.grid(row=5,column=4)
site_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
site_entry.grid(row=6,column=1)
exptime_entry = Entry(windows,width = 20, font=('arial',17,'bold'), bd=2, insertwidth = 2, justify = 'right')
exptime_entry.grid(row=6,column=4)


# Hiển thị
title_label1 = Label(windows, text = "HIỂN THỊ THÔNG TIN VẬT TƯ",font=('arial',25,'bold')).grid(row=8,column=0,columnspan=4)

display_listbox = Listbox(windows)
display_listbox.grid(row=9,column=0,columnspan=5)
'''
impQua_label1 = Label(windows, text = "Số lượng nhập",font=('arial',10,'bold'))
impQua_label1.grid(row=9,column=1)
imptime_label1 = Label(windows, text = "Thời gian nhập",font=('arial',10,'bold'))
imptime_label1.grid(row=9,column=2)
exp_label1 = Label(windows, text = "Hạn sử dụng",font=('arial',10,'bold'))
exp_label1.grid(row=9,column=3)

expQua_label1 = Label(windows, text = "Số lượng xuất",font=('arial',10,'bold'))
expQua_label1.grid(row=9,column=4)
site_label1 = Label(windows, text = "địa điểm",font=('arial',10,'bold'))
site_label1.grid(row=9,column=5)
exptime_label1 = Label(windows, text = "thời gian xuất",font=('arial',10,'bold'))
exptime_label1.grid(row=9,column=6)
inventory_label1 = Label(windows, text = "tồn kho",font=('arial',10,'bold'))
inventory_label1.grid(row=9,column=7)
'''
# Chạy ứng dụng
windows.mainloop()
```

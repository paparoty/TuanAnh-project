```python
'''
Khởi động 1: Viết chương trình in ra tất cả các số chia hết cho cả 5 và 7 trong khoảng từ 1500 đến 2700
'''
```

```python
# vòng lặp while
count = 1500
while count < 2700:
    if not count % 5 and not count % 7: print(count)
    count += 1
print('----------------')
# vòng lặp for
for i in range(1500, 2701):
    if not i % 5 and not i % 7: print(i)
```

```python
'''
Bài 1: Viết chương trình nhập vào chiều cao tam giác vuông và in ra tam giác có hình dạng như sau:
*****
****
***
** 
*
'''
n = int(input("Nhập chiều cao tam giác: "))

# vòng lặp while
count = 1
while count <= n:
    print ("*" * (n - count + 1))
    count += 1
    
print('----------------')

# vòng lặp for
for i in range(n, 0, -1):
    print('*' * i)
```

```python
'''
Bài 2: Viết chương trình nhập vào chiều cao tam giác vuông và in ra tam giác có hình dạng như sau:
* * * * * *
* * * * *
* * * *
* * *
* *
*
'''

n = int(input("Nhập chiều cao tam giác : "))

# vòng lặp while
count = n
while count > 0:
    print("* " * (count - 1) + "*")
    count -= 1

print('----------------')
    
# vòng lặp for
for i in range(n, 0, -1):
    print('* ' * (i - 1) + '*')
```

```python
'''
Bài 3: Viết chương trình nhập vào chiều cao tam giác vuông và in ra tam giác có hình dạng như sau:
* - * - *
* - * -
* - *
* -
*
'''
n = int(input("Nhập chiều cao tam giác: "))

# vòng lặp while
count = n
while count > 0:
    i = 0
    while i < count:
        if i % 2: 
            print('-', end='')
        else: 
            print('*', end='')
        
        if i != count: print(' ', end='')
        i += 1
        
    print()
    count -= 1
    
print('----------------')

# vòng lặp for
for i in range(n, 0, -1):
    for j in range(i):
        print('-' if j % 2 else '*', end='')
        if j != i - 1: print(' ', end='')
    print()
```

```python
'''
Bài 4: Viết chương trình nhập vào một số nguyên n và in ra các số từ 1 đến n với các trường hợp ngoại lệ như sau:
- Nếu số đó chia hết cho 3 thì in "py"
- Nếu số đó chia hết cho 7 thì in "thon"
- Nếu số đó chia hết cho cả 3 và 7 thì in "python"

Ví dụ: n = 25
1
2
py
4
5
py
thon
8
py
10
11
py
13
thon
py
16
17
py
19
20
python
22
23
py
25
'''
n = int(input("Nhập vào một số nguyên: "))

# vòng lặp while
count = 1
while count <= n:
    if not count % 3 and not count % 7:
        print('python')
    elif not count % 3:
        print('py')
    elif not count % 7:
        print('thon')
    else: 
        print(count)
    count += 1

print('----------------')

# vòng lặp for
for i in range(1, n + 1):
    if not i % 3 and not i % 7:
        print('python')
    elif not i % 3:
        print('py')
    elif not i % 7:
        print('thon')
    else: 
        print(i)
```

```python
'''
Bài 5: Viết chương trình nhập vào một số và kiểm tra xem số đó có phải là số nguyên tố hay không
Note: Số nguyên tố là số chỉ chia hết cho 1 và chính nó
'''
n = int(input("Nhập vào một số nguyên: "))

# vòng lặp while
check = True
count = 2
while count < n:
    if not n % count:
        check = False
        break
    count += 1

if check: print(f"{n} là số nguyên tố")
else: print(f"{n} không phải là số nguyên tố")
    
print('----------------')

# vòng lặp for
check = True
for i in range(2, n):
    if not n % i:
        check = False
        break

if check: print(f"{n} là số nguyên tố")
else: print(f"{n} không phải là số nguyên tố")
```

```python
'''
Bài 6: Viết chương trình mô phỏng trò chơi kéo búa bao theo luật chơi như sau:
Quy ước: 1 - búa, 2 - kéo, 3 - bao
Khi trò chơi bắt đầu, yêu cầu người chơi nhập 1 số từ bàn phím đại diện cho kéo, búa, bao
Chương trình tự random ra một lựa chọn từ 1 đến 3.
Nếu người chơi thua thì tiếp tục chơi cho đến khi thắng thì kết thúc chương trình.
Ví dụ:

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 3
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 2
Máy tính chọn: 1
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 2
Người chơi thắng.
'''
from random import randint
win = False
while not win:
    player_choice = int(input("Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): "))
    computer_choice = randint(1, 3)
    print(f"Máy tính chọn: {computer_choice}")
    if player_choice == computer_choice:
        print("Hòa")
    elif player_choice + 1 == computer_choice or player_choice - 2 == computer_choice:
        print("Người chơi thắng")
        win = True
    else:
        print("Người chơi thua.")
```
